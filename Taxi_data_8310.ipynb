```python
## Cleaning the raw data for NYC TAXI FARE PREDICTION Project
```

```python
import pandas as pd
import glob
```

```python
files = glob.glob("Data/*.parquet")
files
```

```python
ff = files
ff
```

```python
dfs = [pd.read_parquet(f) for f in ff]

combined = pd.concat(dfs, ignore_index=True)
```

```python
combined.shape
```

```python
df = combined
#df = pd.read_csv(r"D:\DATA\Project\8420\Taxi_data.csv")
```

```python
df.head(50)
```

```python
print(df['tpep_pickup_datetime'].dtype)
```

```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'], errors='coerce')
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'], errors='coerce')
```

```python
#Confirming if there are trips that with durations being more than a day
df['trip_duration_min'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60
df[(df['trip_duration_min']/60) > 24]
```

```python
#Drop rows where passenger count is NaN
df = df.dropna(subset=['passenger_count'])
```

```python
#Keeping row where passenger counts are not zero
df = df[df['passenger_count'] > 0]
```

```python
print(df['PULocationID'].dtype)
```

```python
#Drop Rows where drop of and pick up locations are outside of New York or not known
df = df[~df['PULocationID'].isin([264,265])]
df = df[~df['DOLocationID'].isin([264,265])]
```

```python
df.dtypes
```

```python
#Removing rows based on their payment type
#0 - rare and mostly a subscription or special fare; doesnt seem to be there in the dataset
#3 - Its free trips and as such generates no fare
#4 - disputed; so could be incorrect or reversed
#5 - unknown; not real case
#6 - voided trips; generates no fare
df = df[~df['payment_type'].isin([0,3,4,5,6])]
```

```python
df['payment_type'].unique()
```

```python
df.sort_values(by='fare_amount', ascending=True).head()
```

```python
#Removing payment type column as it will only contain rows that are paid in cash or by card.
df = df.drop(columns=['payment_type'])
```

```python
#Removing store_and_fwd_flag as its meant for audit purpose and doesnt really effect fare prices:
df = df.drop(columns=['store_and_fwd_flag'])
```

```python
#Removing rows where the ratecodeID  is 99 or just NaN to prevent data quality issues.
df = df.dropna(subset=['RatecodeID'])
df = df[~df['RatecodeID'].isin([99])]
df[df['RatecodeID']==99].count()
```

```python
df.columns
```

```python
df.isna().sum()
```

```python
#Inspecting fare prices
pd.set_option('display.float_format', '{:,.2f}'.format) #Remvoing the exponent format
df['fare_amount'].describe()
```

```python
#There are some extreme outliers......
#To fix this, a fare range will be set from the base fare of $3 to a max of $500
#Why a base of $3? According to NYC.gov, it is the initial charge for a trip.
#Why a max of $500? It is a generous limit but should cover airport, toll and congestion prices

df = df[(df['fare_amount'] >= 3.00) & (df['fare_amount'] <= 500)]
```

```python
#Checking
df['fare_amount'].describe()
```

```python
df.shape
```

```python
df[(df['trip_duration_min']/60) > 24]
#These trips are not realistic as it spans more than a day and as such will be removed. Likely caused by some error, time zone anomalies or just corrupted.
df = df[(df['trip_duration_min']/60) <= 24]
#Ensuring there are no negative or 0 time trips
df = df[df['trip_duration_min'] >0]
```

```python
#We can remove "improvement_surcharge" and "mta_tax" seem to be fixed attributes or have very little deviation and as such can be removed.
df = df.drop(columns=['mta_tax', 'improvement_surcharge'])
```

```python
#"Tip_amount" doesn't really dictate the fare so it's not needed and "total_amount" can cause a data leak.
df = df.drop(columns=['tip_amount', 'total_amount'])
```

```python
df.columns
```

```python
#Since all trips now occur within a single day, we only need one column for the day of the week and one column for the month, based on the pickup time.
#Additionally, a "time of day" category (Morning, Afternoon, Evening and Night) will be created using the pickup time, as the start time has a larger impact on fare than the drop-off time.
```

```python
df['day_of_week'] = df['tpep_pickup_datetime'].dt.day_name()
df['month'] = df['tpep_pickup_datetime'].dt.month_name()
df['year'] = df['tpep_pickup_datetime'].dt.year
df['pickup_hour'] = df['tpep_pickup_datetime'].dt.hour
df['dropoff_hour'] = df['tpep_dropoff_datetime'].dt.hour
```

```python
def time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

df['time_of_day'] = df['pickup_hour'].apply(time_of_day)
```

```python
df.columns
```

```python
df['month'].unique()
```

```python
#Since temporal attributes are created there is no need for the 'tpep_pickup_datetime'and 'tpep_dropoff_datetime' attributes
df = df.drop(columns=['tpep_pickup_datetime', 'tpep_dropoff_datetime'])
df.head()
```

```python
##Some descriptive stats of numerical variable to find outliers
num = df.select_dtypes(include = ['int64','float64']).columns
df[num].describe()
```

```python
#Concerns:
#"cbd_congestion_fee" has negative values
#"trip_distance" has 0 distance trips and a max trip of 48,199.11 which doesnt seem right....
#Even after removing trips with a duration of more than 24 hour, there are some trips that take a whole day

#To prevent issues, these outliers will be removed .....

#Duration
df = df[(df['trip_duration_min'] >= 5) & (df['trip_duration_min'] <= 150)]

#Distance
df = df[(df['trip_distance'] > 0) & (df['trip_distance'] <= 100)]

#These limits are alot more realistic and a bit too generous but should pose no problems.

#The raw fees can also be converted to flags to for potential exploration.

# Flag for airport trips
df['is_airport_trip'] = (df['Airport_fee'] > 0).astype(int)

# Flag for CBD congestion
df['is_cbd_congestion'] = (df['cbd_congestion_fee'] > 0).astype(int)

# Flag for extra charges
df['has_extra'] = (df['extra'] > 0).astype(int)

# Flag for tolls
df['has_toll'] = (df['tolls_amount'] > 0).astype(int)
```

```python
#Removing the raw fee and unnecessary columns
df = df.drop(columns=['Airport_fee', 'cbd_congestion_fee', 'extra', 'tolls_amount', 'congestion_surcharge', 'year', 'VendorID', 'PU_Borough', 'dropoff_hour',	'pickup_hour','month'])
```
